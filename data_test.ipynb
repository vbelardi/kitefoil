```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
file_path = r"d:\Universitaet\Lausanne\Cours\MA3\Formula_Kite\Kitefoil 2.0 & 3.0\Code\kitefoil\final_test_2.csv"
data = pd.read_csv(file_path, sep=";")
data.head()
```

```python
# Calculate the time difference between each sample
data["Time Difference (ms)"] = data["CPU Timestamp (ms)"].diff()
data["Time Difference GPS"] = data["CPU Timestamp (ms)"].diff()

# Plot a histogram of the time differences
plt.figure(figsize=(10, 6))
plt.hist(
    data["Time Difference (ms)"].dropna(), bins=50, color="blue", edgecolor="black"
)
plt.title("Histogram of Time Differences Between Samples")
plt.xlabel("Time Difference (ms)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Convert the 'Timestamp GPS' column to datetime
data["Timestamp GPS"] = pd.to_datetime(data["Timestamp GPS"], errors="coerce")

# Filter rows where 'Timestamp GPS' changes
data_filtered = data[data["Timestamp GPS"] != data["Timestamp GPS"].shift()]

# Calculate the time difference in milliseconds for the filtered 'Timestamp GPS' column
data_filtered["Time Difference GPS (ms)"] = (
    data_filtered["Timestamp GPS"].diff().dt.total_seconds() * 1000
)

# Plot a histogram of the time differences for 'Timestamp GPS'
plt.figure(figsize=(10, 6))
plt.hist(
    data_filtered["Time Difference GPS (ms)"].dropna(),
    bins=50,
    color="green",
    edgecolor="black",
)
plt.title("Histogram of Time Differences Between GPS Timestamps (Value Changes Only)")
plt.xlabel("Time Difference (ms)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Convert the 'Timestamp GPS' column to datetime
data["Timestamp GPS"] = pd.to_datetime(data["Timestamp GPS"], errors="coerce")

# Filter rows where 'Timestamp GPS' changes
data_filtered = data[data["Timestamp GPS"] != data["Timestamp GPS"].shift()]

# Calculate the time difference for 'CPU Timestamp (ms)' and 'Timestamp GPS'
data_filtered["CPU Time Difference (ms)"] = data_filtered["CPU Timestamp (ms)"].diff()
data_filtered["GPS Time Difference (ms)"] = (
    data_filtered["Timestamp GPS"].diff().dt.total_seconds() * 1000
)

# Compare the time differences
data_filtered["Time Difference Delta (ms)"] = (
    data_filtered["CPU Time Difference (ms)"]
    - data_filtered["GPS Time Difference (ms)"]
)

# Drop rows with NaN values (first row after diff will have NaN)
data_filtered = data_filtered.dropna()

# Display the comparison
# print(data_filtered[["CPU Time Difference (ms)", "GPS Time Difference (ms)", "Time Difference Delta (ms)"]])
data_filtered
```

```python
# Plot the comparison of time differences
plt.figure(figsize=(10, 6))
plt.plot(
    data_filtered["CPU Time Difference (ms)"],
    label="CPU Time Difference (ms)",
    marker="o",
)
plt.plot(
    data_filtered["GPS Time Difference (ms)"],
    label="GPS Time Difference (ms)",
    marker="x",
)
plt.title("Comparison of Time Differences (CPU vs GPS)")
plt.xlabel("Index")
plt.ylabel("Time Difference (ms)")
plt.legend()
plt.grid(True)
plt.show()

# Plot the delta (difference between the two time differences)
plt.figure(figsize=(10, 6))
plt.plot(
    data_filtered["Time Difference Delta (ms)"],
    label="Time Difference Delta (ms)",
    color="red",
    marker="o",
)
plt.axhline(0, color="black", linestyle="--", linewidth=1)
plt.title("Delta Between CPU and GPS Time Differences")
plt.xlabel("Index")
plt.ylabel("Delta (ms)")
plt.legend()
plt.grid(True)
plt.show()
```
